```python
import requests
from bs4 import BeautifulSoup

# URL of the Scopus search results page
url = 'https://www.scopus.com/results/results.uri?sort=cp-f&src=s&st1=english&sid=284d9d20df10bdcfb93bc932b9e6d819&sot=b&sdt=cl&sl=17&s=LANGUAGE%28english%29&origin=resultslist&editSaveSearch=&yearFrom=2010&yearTo=2025&sessionSearchId=284d9d20df10bdcfb93bc932b9e6d819&limit=100&cluster=scosubjabbr%2C%22MEDI%22%2Ct%2Bscosubtype%2C%22ar%22%2Ct%2Bscoaffilctry%2C%22Thailand%22%2Ct'

# Send a GET request to the URL
response = requests.get(url)

print(response.text)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find all document items
document_items = soup.find_all('div', class_='documentItem')

# Initialize a counter to limit to the first 100 documents
count = 0

for item in document_items:
    # Get the document title
    title = item.find('a', class_='titleLink').text.strip()
    
    # Get the document authors
    authors = item.find('div', class_='authorGroup').text.strip()
    
    # Get the document publication date
    date = item.find('span', class_='text').text.strip()
    
    # Get the document citations
    citations = item.find('span', class_='citedByCount').text.strip()
    
    # Get the document abstract (if available)
    abstract = item.find('div', class_='abstractEllipsis')
    if abstract:
        abstract = abstract.text.strip()
    
    # Get the document country (assuming it's in the author affiliations)
    country = None
    affiliations = item.find_all('li', class_='affiliationListItem')
    for affiliation in affiliations:
        if 'Thailand' in affiliation.text:
            country = 'Thailand'
            break
    
    # Print the document information
    print(f"Title: {title}")
    print(f"Authors: {authors}")
    print(f"Publication Date: {date}")
    print(f"Citations: {citations}")
    print(f"Country: {country}")
    if abstract:
        print(f"Abstract: {abstract}")
    print("\n")
    
    count += 1
    if count >= 100:
        break
```
